# Como pegar pessoas com interesse em desenvolvimento e design?


**Primeira Estratégia**
1.   Pegar todas os uids das vagas de Design e Desenvolvimento **uid_vagas**
2.   Pegar todos os uid dos inscritos das vagas **uid_candidatos**
3.   Pegar todos os emails dos inscritos **emails** + **nomes**
4.   Criar um dataframe com os dados, com colunas **nome** e **email**
5.   Salvar em excel **newtab_publico.xlsx**





**Coleções envolvidas** 

*   Coleção de vagas: **Vagas** (coleções antigas de vagas não tinham área)
*   Coleção de inscrições: **inscricoesCandidatos**
*   Coleção de candidatos: **candidatos**


```python
vacancies = db.collection(u"Vagas")
```

```python
example = {'areas': [{'value': 'comercial', 'subArea': '', 'label': 'Comercial'},{'value': 'desenvolvedor', 'subArea': '', 'label': 'Desenvolvedor'},{'value': 'design', 'subArea': '', 'label': 'Design'}]}
```

```python
def checkArea (areaOne, areaTwo, dictionary):
  return dictionary['value'] == areaOne or dictionary['value'] == areaTwo
```

```python
for element in example['areas']:
  print(checkArea('desenvolvedor','design',element))
```

```python
uid_vagas = []
```

```python
for doc in vacancies.get():
  doc_to_dict = doc.to_dict()
  uid = doc.id
  for area in doc_to_dict['areas']:
    if checkArea('desenvolvedor','design', area):
      uid_vagas.append(uid)
      break
    else:
      None
      
```
